```python
from sklearn.model_selection import cross_val_predict
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import svm
```

**Importando dataset Iris**

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
#importando a função train_test_split para separa conjunto de treino e teste 70/30
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state =0)
```

```python
clf = svm.SVC(kernel='linear', C=1).fit(X_train,y_train)
```

```python
clf.score(X_test, y_test)
```

```python
#Funçao que retorna a acuracia após fazer uma validação cruzada
def Acuracia(clf,X,y):
    resultados= cross_val_predict(clf, X, y, cv=5)
    return metrics.accuracy_score(y, resultados)
```

```python
Acuracia(clf, iris.data, iris.target)
```

```python
clf = svm.SVC(kernel='rbf', C=1).fit(X_train,y_train)
```

```python
clf.score(X_test, y_test)
```

```python
Acuracia(clf, iris.data, iris.target)
```

```python
clf = svm.SVC(kernel='poly', C=1).fit(X_train,y_train)
```

```python
clf.score(X_test, y_test)
```

```python
Acuracia(clf, iris.data, iris.target)
```
